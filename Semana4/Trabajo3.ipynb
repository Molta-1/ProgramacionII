## 2. Usando for(...) en Python muestre los números del 1 al 100 que terminan en 3.

```python
num = 3
limite = 100

for indice in range(limite):
   if indice == num:
      print(num)
      num = num + 10
   if num >= 100:
      break  
```

## 3. Mediante un ciclo, calcule la sumatoria de los números enteros multiplos de 13, comprendidos entre el 1 y el 200.

```python
limite = 200
sumatoria = 0

for indice in range(limite):
    if indice % 13 == 0:
        sumatoria = sumatoria + indice
        print(sumatoria)
    if indice >= 200:
        break    
```

## 6. Programe en Python una función que genera 200 numeros al azar entre 1 y 500 y luego calcula cuantos están entre el 50 y 450, ambos inclusive.


```python
def numeros():
    lista = [rd.randint(1, 500) for _ in range(200)]
    contador = 0
    for i in lista:
        if i >= 50 and i <= 450:
            contador = contador + 1
    return(contador)        

print("La cantidad de números entre 50 y 450 es:",numeros())
```

## 7.Desarrolle una funcion que reciba dos numeros enteros a y b, tal que, en un diccionario retorne el Máximo Común Divisor (MCD) y el mınimo comun multiplo (mcm). Para calcular el MCD puede utilizar la funcion gcd del paquete math.

```python

def calcular_mcd_mcm(a, b):
    mcd = math.gcd(a, b)
    mcm = abs(a * b) // mcd if mcd != 0 else 0
    
    return {
        "MCD": mcd,
        "mcm": mcm
    }

resultado = calcular_mcd_mcm(12, 18)
print(resultado)
```

## Desarrolle una función que recibe una matriz cuadrada A de tamaño n × n y retorna su transpuesta (utilizando for).

```python
matriz = np.array([[0, 20, 6],
                   [4, 12, 48],
                   [-8, -5, -16]])

def matriz_TRANS():
    linea1 = []
    linea2 = []
    linea3 = []

    for x in np.nditer(matriz[:, 0]):
        linea1.append(x)

    for x in np.nditer(matriz[:, 1]):
        linea2.append(x)

    for x in np.nditer(matriz[:, 2]):
        linea3.append(x)

    nueva_matriz = np.array([linea1, linea2, linea3])  
    return nueva_matriz

print(matriz_TRANS())
```

## 9. Desarrolle una funcion que recibe un DataFrame y un nombre de columna y que retorne un diccionario con los valores de la  media, mínimo, máximo y mediana de dicha columna del dataframe. Verifique la correctitud de esta funcion usando la tabla Players1.csv y la columna minutes, que hace referencia a los minutos de juego.

```python
players = pd.read_csv('Players1.csv')
print(players)

def diccionario(df, ic):
    columna_seleccionada = df.iloc[:, ic].values

    resultado = {
        "media": float(np.mean(columna_seleccionada)),
        "mediana": float(np.median(columna_seleccionada)),
        "minimo": float(np.min(columna_seleccionada)),
        "maximo": float(np.max(columna_seleccionada))
    }
    
    return resultado

estadisticas = diccionario(players, 3)
print(estadisticas)
```

## 10. Desarrolle una función que recibe un DataFrame y dos numeros de columna y que retorna la correlacion entre esas dos variables. Verifique la correctitud de esta funcion usando la tabla Players1.csv y las columnas 5 y 6, que hacen referencia a las variables shots y passes.

```python
def calcular_correlacion(df, ic1, ic2):
    columna1 = df.iloc[:, ic1]
    columna2 = df.iloc[:, ic2]

    correlacion = columna1.corr(columna2)
    return correlacion

resultado = calcular_correlacion(players, 4, 5)

print(resultado)
```
